```python
import matplotlib.pyplot as plt
import numpy as np
```

### Forward Euler Method

```python
x_0 = 0
x_n = 0.25
h = 0.05

# f(x,y) = -xy
# y(0) = 1

y_vals = [1] # initial value of y at x = 0
x_vals = [0, 0.05, 0.1, 0.15, 0.2, 0.25]#, 0.3, 0.35, 0.4, 0.45, 0.5] #np.linspace(0,25)#
i = 0 # iterator for y_i
print("Forward Euler Method")
for x in x_vals:
    y_vals.append(y_vals[i]-h*x*y_vals[i])
    print(x,"\t:",y_vals[i])
    i+=1

print("\nAnalytical Solution")
y_actual = []
for x in x_vals:
    y_actual.append(np.power(np.e, (-x**2)/2))
    print(x,"\t:",y_actual[-1])

print("\nError Percentage:")
i=0
while i<len(y_vals)-1:
    print("Iteration - ", i, "Error: ", np.abs((y_actual[i]-y_vals[i]))*100, "%")
    i+=1
```

```python
plt.style.use("ggplot")
plt.grid(True)
plt.plot(x_vals, y_vals, 'o--', label="Forward Euler")
plt.plot(x_vals, y_actual, 'o-', label="Analytical Solution")
plt.legend(loc='best')
plt.savefig("Euler.png", dpi=300)
```

### Huen's Method

```python

x_0 = 0
x_n = 0.25
h = 0.05

# f(x,y) = -xy
# y(0) = 1
def slope(x, y, pos=True):
    if pos:
        return x*y
    else:
        return -1*x*y

y_vals_huen = [1] # initial value of y at x = 0
x_vals = [0, 0.05, 0.1, 0.1, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5] #np.linspace(0,25)#

for x in x_vals:
    s1 = slope(x, y_vals_huen[-1]
    s2 = -1*(x+h * (y_vals_huen[-1] + s1*h))
    s_avg = (s1 + s2)/2
    #print("s1: ", s1, "\ts2: ", s2, "\ts3: ",s_avg)
    #print(x+h, (y_vals[-1] + s1*h))
    y_vals_huen.append(y_vals_huen[-1]+h*(s_avg))
    #print(x,"\t:", y_vals[-1])
```
